```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("BlackFriday.csv")
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
l = []
for i in range(len(df)):
    l.append(df['Gender'][i] + "_" + str(df['Marital_Status'][i]))
    
df['Marital_Gender'] = l
```

```python
df.head()
```

```python
sns.countplot(x = df['Marital_Gender']).set(title = "Merital Gender Purchase Data")
```

```python
sns.countplot(x = df['Age'], hue = df['Marital_Gender']).set(title = "Merital Gender Purchase Data A/C TO AGE")
```

```python
sns.countplot(x = df['Age'], hue = df['Product_Category_1']).set(title = "Merital Gender Purchase Data A/C TO PRODUCT CATEGORY")
```

```python
sns.countplot(x = df['Stay_In_Current_City_Years'], hue = df['Marital_Gender']).set(title = "Merital Gender Purchase Data A/C TO STAY IN CITY YEARS ")
```

```python
sns.countplot(x = df['City_Category'], hue = df['Marital_Gender']).set(title = "Merital Gender Purchase Data A/C TO CITY CATEGORY ")

```
