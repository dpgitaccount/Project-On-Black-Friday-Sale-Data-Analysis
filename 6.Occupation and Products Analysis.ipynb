```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("BlackFriday.csv")
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years', hue = 'Gender', data = df ).set(title = 'EXPERIENCE OF PEOPLE')
```

```python
df.groupby('Stay_In_Current_City_Years').size().plot(kind = 'pie', autopct = '%.1f')
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years', hue = 'Marital_Status', data = df ).set(title ='MARITAL STATUS OF PEOPLES')
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years', hue = 'City_Category', data = df ).set(title = 'PEOPLES CITY CATEGORY')
```

```python
sns.countplot(x = 'City_Category', hue = 'Age', data = df ).set(title = 'AGE OF PEPLE ACCORDING TO CITY_CATEGORY')
```

```python
df['Purchase'] = pd.to_numeric(df['Purchase'], errors='coerce')
df.groupby('Stay_In_Current_City_Years')['Purchase'].apply(lambda x: pd.to_numeric(x, errors='coerce').sum()).plot(kind='bar', figsize=(12, 6), title='TOTAL AMOUNT SPEND BY CITY CATEGORY')
```

```python
df['Purchase'] = pd.to_numeric(df['Purchase'], errors='coerce')
df.groupby('Stay_In_Current_City_Years')['Purchase'].apply(lambda x: pd.to_numeric(x, errors='coerce').mean()).plot(kind='bar', figsize=(12, 6), title='AVERAGE AMOUNT SPEND BY CITY CATEGORY')
```

```python
sns.countplot(x = df['Occupation']).set(title = "OCCUPATION OF PEOPLES")
```

```python
df.groupby('Occupation').size().sort_values().plot(kind = 'bar').set(title = "OCCUPATION OF PEOPLES")
```

```python
df['Purchase'] = pd.to_numeric(df['Purchase'], errors='coerce')
df.groupby('Occupation')['Purchase'].apply(lambda x: pd.to_numeric(x, errors='coerce').sum()).sort_values().plot(kind='bar', figsize=(12, 6), title='TOTAL MONEY SPEND A/C TO OCCUPATION')
```

```python
df['Purchase'] = pd.to_numeric(df['Purchase'], errors='coerce')
df.groupby('Occupation')['Purchase'].apply(lambda x: pd.to_numeric(x, errors='coerce').mean()).sort_values().plot(kind='bar', figsize=(12, 6), title='AVERAGE MONEY SPEND A/C TO OCCUPATION')
```

```python
df['Purchase'] = pd.to_numeric(df['Purchase'], errors='coerce')
df.groupby('Occupation')['Purchase'].apply(lambda x: pd.to_numeric(x, errors='coerce').mean()).sort_values()
```

```python
sns.countplot(x = 'Occupation', hue = 'Marital_Status', data = df).set(title = 'MARITAL STATUS A/C TO OCCUPATION')
```

```python
sns.countplot(x = 'Occupation', hue = 'Gender', data = df).set(title = 'GENDER A/C TO OCCUPATION')
```

```python
df.groupby('Occupation').nunique()['Product_ID'].sort_values().plot(kind = 'bar').set(title = 'UNIQUE PRODUCTS ')
```

```python
df.groupby('Occupation').nunique()['Product_ID'].plot(kind = 'pie', autopct = '%.1f').set(title = 'UNIQUE PRODUCTS ')
```

```python
df.groupby('Product_Category_1').size().sort_values().plot(kind = 'bar').set(title = 'PRODUCT CATEGORY ')
```

```python
df['Purchase'] = pd.to_numeric(df['Purchase'], errors='coerce')
df.groupby('Product_Category_1')['Purchase'].apply(lambda x: pd.to_numeric(x, errors='coerce').mean()).sort_values().plot(kind = 'bar').set(title = 'AVERAGE AMOUNT SPEND')
```

```python
df['Purchase'] = pd.to_numeric(df['Purchase'], errors='coerce')
df.groupby('Product_ID')['Purchase'].apply(lambda x: pd.to_numeric(x, errors='coerce').sum()).nlargest(10).sort_values().plot(kind = 'bar').set(title = 'TOP 10 PRODUCTS')
```
