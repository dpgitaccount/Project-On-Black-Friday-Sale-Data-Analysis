```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("BlackFriday.csv")
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
df.groupby('Age').size().plot(kind = 'bar', figsize = (12,6), title = 'PURCAHSE DISTRIBUTION BY AGE')
```

```python
sns.set(rc = {'figure.figsize' : (12,6)})
sns.countplot(x = "Age" , hue = 'Gender', data = df).set(title='PURCHASE DISTRIBUTION BY AGE & ACCORDING TO GENDER')
```

```python
sns.set(rc = {'figure.figsize' : (12,6)})
sns.countplot(x = "Gender" , hue = 'Age', data = df).set(title='PURCHASE DISTRIBUTION BY AGE & ACCORDING TO GENDER')
```

```python
sns.set(rc = {'figure.figsize' : (6,6)})
sns.countplot(x = "Gender" , hue = 'Marital_Status', data = df).set(title='MARITAL STATUS OF MALE AND FEMALE CUSTOMER')
```

```python
sns.countplot(x = df['City_Category']).set(title='OVERALL PURCHASED BY CITY')
```

```python
sns.countplot(x= 'City_Category', hue = 'Age', data = df).set(title='CITY CATEGORY')
```

```python
sns.countplot(x= 'Marital_Status', hue = 'City_Category', data = df).set(title='MARITAL STATUS ACCORDING TO CITY CATEGORY')
```

```python
df.groupby('City_Category').size().plot(kind = 'pie', autopct = '%0.1f')
```

```python
sns.countplot(x= 'City_Category', hue = 'Gender', data = df).set(title=' PURCHASED AMOUNT CITY CATEGORY ACCORDING TO GENDER')
```

```python
df['Purchase'] = pd.to_numeric(df['Purchase'], errors='coerce')

# Group by 'Age', calculate mean 'Purchase', and plot
# Filter out non-numeric values before calculating the mean
df.groupby('City_Category')['Purchase'].apply(lambda x: pd.to_numeric(x, errors='coerce').sum()).plot(kind='pie', autopct = '%0.1f' , figsize=(12, 6), title='AMOUNT SPEND ACCORDING TO CITY CATEGORY')
```

```python
df['Purchase'] = pd.to_numeric(df['Purchase'], errors='coerce')
df.groupby('City_Category')['Purchase'].apply(lambda x: pd.to_numeric(x, errors='coerce').mean()).plot(kind='pie', autopct = '%0.1f' , figsize=(12, 6), title='AVERAGE AMOUNT SPEND BY CITY CATEGORY')
```
