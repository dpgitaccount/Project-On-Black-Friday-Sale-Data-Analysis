```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("BlackFriday.csv")
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
## Purchase Distibution On Age of Customers
df.groupby('Age').size().plot(kind = 'bar', figsize = (12,6), title = 'PURCAHSE DISTRIBUTION BY AGE')
```

```python
lst = []
for i in df['Age'].unique():
    lst.append([i, df[df['Age'] == i]['Product_ID'].nunique()])
data = pd.DataFrame(lst, columns = ['Age', 'Products'])
data
```

```python
data.plot.bar(x = 'Age', figsize = (8,6), title = "UNIQUE PRODUCTS ACCORDING TO AGE OF CUSTOMERS")
```

```python
df.groupby('Age').sum()['Purchase'].plot(kind = 'bar', figsize = (12,6), title = 'AMOUNT SPEND BY AGE')
```

```python
df['Purchase'] = pd.to_numeric(df['Purchase'], errors='coerce')

# Group by 'Age', calculate mean 'Purchase', and plot
# Filter out non-numeric values before calculating the mean
df.groupby('Age')['Purchase'].apply(lambda x: pd.to_numeric(x, errors='coerce').mean()).plot(kind='pie', autopct = '%0.1f' , figsize=(12, 6), title='AVERAGE AMOUNT SPEND BY AGE')
```

```python
df.groupby('Marital_Status').size().plot(kind = 'pie', autopct = '%0.1f', title = 'MARITAL STATUS OF MALE AND FEMALE CUSTOMERS')
```

```python
df.groupby('Gender').size().plot(kind = 'pie', autopct = '%0.1f', title = 'GENDER STATUS OF CUSTOMERS')
```
